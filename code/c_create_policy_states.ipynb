```python
import pandas as pd
```

```python
# load the data with the states and counties and their fips
counties = pd.read_csv('https://raw.githubusercontent.com/kjhealy/fips-codes/master/county_fips_master.csv', encoding = 'latin-1')
```

```python
# create fips code for the states
counties['county_fips'] = counties['fips'].astype(str).str.rjust(5, '0')
```

```python
counties.head()
```

```python
# create the state fips for each state
counties['state_fips'] = counties['county_fips'].str[:2].str.ljust(5, '0')
```

```python
# create a list of the variables I will keep
keep_list = ['state_name', 'state_fips', 'county_name', 'county_fips']
data = counties[keep_list]
```
